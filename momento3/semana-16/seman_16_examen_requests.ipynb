```python
import pandas as pd
import matplotlib.pyplot as plt
import requests

## URL de la API
url = "https://api.covidtracking.com/v1/us/daily.json"

## Realizamos la peticin y guardamos la respuesta como un JSON
response = requests.get(url).json()

## Convertimos la respuesta a un DataFrame de Pandas
df = pd.DataFrame(response)

## Convertimos la columna 'date' a tipo datetime
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

## Graficamos el nmero de casos positivos
plt.plot(df['date'], df['positive'])
plt.xlabel('Fecha')
plt.ylabel('Casos positivos')
plt.title('Casos positivos de COVID-19 en EE. UU.')
plt.show()

## Graficamos el nmero de muertes
plt.plot(df['date'], df['death'])
plt.xlabel('Fecha')
plt.ylabel('Muertes')
plt.title('Muertes de COVID-19 en EE. UU.')
plt.show()

```
